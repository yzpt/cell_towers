```python
gdf.columns = ['radio', 'mcc', 'net', 'area', 'cell', 'unit', 'lon', 'lat', 'range', 'samples', 'changeable', 'created', 'updated', 'averageSignal']
gdf.sample(10)
```

```python
# Create geometry from lon and lat
gdf['geometry'] = gpd.points_from_xy(gdf['lon'], gdf['lat'])
gdf.head()
```

```python
gdf = gpd.GeoDataFrame(gdf, geometry='geometry')
type(gdf)
```

```python
# Set the GeoDataFrame CRS (e.g., WGS84/espg=4326 for GPS data)
gdf.set_crs(epsg=4326, inplace=True)
```

```python
# export gdf to geojson file
gdf.to_file('data/208.geojson', driver='GeoJSON')
# ~ 150 Mo
```
